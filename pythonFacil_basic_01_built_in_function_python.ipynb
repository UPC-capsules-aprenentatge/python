**List to Tuple and Vice Versa**

```python
# Define list and tuple

lst = [1, 2, 3]
tup = (4, 5, 6)
# Convert and print
lst_to_tup = tuple(lst)
print(lst_to_tup) # Output: (1, 2, 3)
print(type(lst_to_tup))

tup_to_lst = list(tup)
print(tup_to_lst) # Output: [4, 5, 6]
print(type(tup_to_lst))

print(tup_to_lst[0])
print(lst_to_tup[0])

print(len(tup_to_lst))
print(len(lst_to_tup))

print(min(tup_to_lst))
print(min(lst_to_tup))

```

**Dicctionaries**

```python
# Dictionary example in Python

my_dict = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Accessing values
print(my_dict["name"])   # John
print(my_dict["age"])    # 30
print(my_dict["city"])   # New York

# Adding a new key-value pair
my_dict["job"] = "Engineer"
print(my_dict)

# Modifying a value
my_dict["age"] = 31
print(my_dict)

# Removing a key
del my_dict["city"]
print(my_dict)

# Checking if a key exists
print("name" in my_dict)  # True

# Getting all keys
print(my_dict.keys())

# Getting all values
print(my_dict.values())

# Getting all items (key-value pairs)
print(my_dict.items())
```

Sets

```python
# Create set with duplicates
nums = {1, 2, 2, 3}
print(nums)  # >>> {1, 2, 3}

# Sets automatically remove duplicates
print("Initial set:", nums)

# Sets are mutable (you can modify them)
nums.add(4)
print("After add(4):", nums)

nums.remove(2)
print("After remove(2):", nums)

# Membership test
print(3 in nums)   # True
print(10 in nums)  # False

# Adding multiple elements (using update)
nums.update([3, 5, 6, 7])
print("After update:", nums)

# Set operations example
a = {1, 2, 3}
b = {3, 4, 5}

print("Union:", a | b)        # {1, 2, 3, 4, 5}
print("Intersection:", a & b)  # {3}
print("Difference 1:", a - b)    # {1, 2}
print("Difference 2:", b -a )    # {4, 5}
```
