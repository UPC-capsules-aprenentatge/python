**Numerical functions**

```python
# abs() - returns absolute value
print(abs(-10))  # 10

# round() - rounds a number
print(round(3.14159, 2))  # 3.14

# max() - returns the largest value
print(max(5, 10, 3))  # 10

# min() - returns the smallest value
print(min(5, 10, 3))  # 3

# sum() - adds all elements in an iterable
numbers = [1, 2, 3, 4]
print(sum(numbers))  # 10

# pow() - power (x to the power of y)
print(pow(2, 3))  # 8

# sorted() - returns a sorted list
nums = [3, 1, 4, 2]
print(sorted(nums))  # [1, 2, 3, 4]
```

**Sequence Functions**

```python
# Sequence Functions in Python (no lambda, no reduce)

# list() - creates a list
print(list("hello"))  # ['h', 'e', 'l', 'l', 'o']

# tuple() - creates a tuple
print(tuple([1, 2, 3]))  # (1, 2, 3)

# range() - generates a sequence of numbers
print(list(range(5)))  # [0, 1, 2, 3, 4]

# sorted() - returns a sorted sequence
nums = [3, 1, 4, 2]
print(sorted(nums))  # [1, 2, 3, 4]

# reversed() - reverses a sequence
print(list(reversed([1, 2, 3])))  # [3, 2, 1]

# enumerate() - adds index to elements
fruits = ["apple", "banana", "cherry"]
print(list(enumerate(fruits)))
# [(0, 'apple'), (1, 'banana'), (2, 'cherry')]

# zip() - combines sequences
names = ["Anna", "John"]
ages = [20, 30]
print(list(zip(names, ages)))
# [('Anna', 20), ('John', 30)]

# map() - applies a function to all elements (no lambda)
def double(x):
    return x * 2

nums = [1, 2, 3]
print(list(map(double, nums)))
# [2, 4, 6]

# filter() - filters elements by condition (no lambda)
def is_even(x):
    return x % 2 == 0

nums = [1, 2, 3, 4]
print(list(filter(is_even, nums)))
# [2, 4]

# slice() - creates a slice object
text = "Python"
s = slice(1, 4)
print(text[s])  # 'yth'
```

**String Functions in Python**

```python


# str() - converts a value to string
x = 10
print(str(x))  # "10"

# chr() - returns character from ASCII code
print(chr(65))  # 'A'

# ord() - returns ASCII code from character
print(ord('A'))  # 65

# len() - returns length of string
text = "Python"
print(len(text))  # 6

# format() - formats strings
name = "Anna"
age = 25
print("My name is {} and I am {} years old".format(name, age))

# find() - finds position of substring (returns -1 if not found)
text = "Hello Python"
print(text.find("Python"))  # 6

# index() - like find() but raises error if not found
print(text.index("Hello"))  # 0

# replace() - replaces text (included once, duplicates removed)
text = "I like Java"
print(text.replace("Java", "Python"))  # I like Python

# split() - splits string into list (included once, duplicates removed)
text = "apple,banana,orange"
print(text.split(","))  # ['apple', 'banana', 'orange']

# join() - joins elements into a string
words = ["Hello", "world"]
print(" ".join(words))  # "Hello world"
```

**Collection Functions**

```python
# set() - creates a set (unique elements)
print(set([1, 2, 2, 3]))  # {1, 2, 3}

# frozenset() - immutable set
fs = frozenset([1, 2, 3])
print(fs)  # frozenset({1, 2, 3})

# dict() - creates a dictionary
d = dict(name="Anna", age=25)
print(d)  # {'name': 'Anna', 'age': 25}

# list() - creates a list
print(list("abc"))  # ['a', 'b', 'c']

# tuple() - creates a tuple
print(tuple([1, 2, 3]))  # (1, 2, 3)

# bool() - converts value to True/False
print(bool(0))   # False
print(bool(10))  # True

# any() - True if at least one element is True
print(any([False, False, True]))  # True

# all() - True if all elements are True
print(all([True, True, False]))  # False

# hash() - returns hash value of an object
print(hash("hello"))  # integer hash value (changes between runs sometimes)
```

**Basic Mathematical Functions in Python**

```python


# abs() - absolute value
print(abs(-10))  # 10

# round() - rounds a number
print(round(3.14159, 2))  # 3.14

# pow() - power (x raised to y)
print(pow(2, 3))  # 8

# min() - smallest value
print(min(5, 10, 3))  # 3

# max() - largest value
print(max(5, 10, 3))  # 10

# sum() - adds all elements in a list
numbers = [1, 2, 3, 4]
print(sum(numbers))  # 10

# divmod() - returns quotient and remainder
print(divmod(10, 3))  # (3, 1)
```

**Iterable Functions in Python**

```python


# iter() - creates an iterator from an iterable
nums = [1, 2, 3]
it = iter(nums)
print(next(it))  # 1

# next() - gets next item from iterator
print(next(it))  # 2
print(next(it))  # 3

# enumerate() - adds index to elements
fruits = ["apple", "banana", "cherry"]
print(list(enumerate(fruits)))
# [(0, 'apple'), (1, 'banana'), (2, 'cherry')]

# zip() - combines iterables
names = ["Anna", "John"]
ages = [20, 30]
print(list(zip(names, ages)))
# [('Anna', 20), ('John', 30)]

# map() - applies a function to each element
def square(x):
    return x * x

nums = [1, 2, 3]
print(list(map(square, nums)))
# [1, 4, 9]

# filter() - filters elements by condition
def is_even(x):
    return x % 2 == 0

nums = [1, 2, 3, 4]
print(list(filter(is_even, nums)))
# [2, 4]

# all() - True if all elements are True
print(all([True, True, False]))  # False

# any() - True if at least one element is True
print(any([False, False, True]))  # True
```

**File Functions in Python**

```python


# open() - opens a file
file = open("example.txt", "w")  # write mode

# write() - writes a string to a file
file.write("Hello world\n")
file.write("Python is awesome\n")

# writelines() - writes a list of strings
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
file.writelines(lines)

# close() - closes the file
file.close()

# reopen file in read mode
file = open("example.txt", "r")

# read() - reads entire file
content = file.read()
print(content)

# close and reopen for next examples
file.close()
file = open("example.txt", "r")

# readline() - reads one line
print(file.readline())  # Line 1

# readlines() - reads all lines into a list
print(file.readlines())

# seek() - moves cursor to a position
file.seek(0)
print("seek(0)", file.read())

# tell() - shows current cursor position
file.seek(5)
print("seek(5)", file.tell())

file.close()
```

**OS Functions**

```python
import os

# os.getcwd() - gets current working directory
print(os.getcwd())

# os.listdir() - lists files and folders in current directory
print(os.listdir())

# os.mkdir() - creates a new folder
os.mkdir("test_folder")
print("Folder created")

# os.path.exists() - checks if a file/folder exists
print(os.path.exists("test_folder"))  # True

# os.rename() - renames a file or folder
os.rename("test_folder", "new_folder")
print("Folder renamed")

# create a file
with open("new_folder/temp0.txt", "w") as f:
    f.write("temporary file")

# os.remove() - deletes a file
# First create a file to delete
with open("temp.txt", "w") as f:
    f.write("temporary file")


print(os.path.exists("temp.txt"))  # True
os.remove("temp.txt")
print(os.path.exists("temp.txt"))  # False
```

**Date and Time Functions in Python**

```python
# Date and Time Functions in Python

import datetime

# datetime.now() - current date and time
print(datetime.datetime.now())

# date.today() - current date
print(datetime.date.today())

# time() - creates a time object
t = datetime.time(14, 30, 45)
print(t)  # 14:30:45

# strftime() - formats date/time as string
now = datetime.datetime.now()
print(now.strftime("%Y-%m-%d %H:%M:%S"))

# strptime() - converts string to datetime
date_str = "2026-06-28 12:30:00"
dt = datetime.datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
print(dt)

# timedelta() - time difference
delta = datetime.timedelta(days=5, hours=3)
print("delta", delta)

# using timedelta with dates
future_date = datetime.datetime.now() + delta
print("future_date", future_date)
```

**Random Functions in Python**

```python


import random

# random() - random float between 0 and 1
print(random.random())

# randint() - random integer between two values
print(random.randint(1, 10))

# uniform() - random float between two values
print(random.uniform(1.0, 10.0))

# choice() - randomly selects one element
fruits = ["apple", "banana", "cherry"]
print(random.choice(fruits))

# shuffle() - randomly shuffles a list
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)

# sample() - picks random unique elements
print(random.sample(numbers, 2))
```
