```python
class ItemComanda:
    def __init__(self, id_producte, quantitat, preu):
        self.id_producte = id_producte
        self.quantitat = quantitat
        self.preu = preu

    def subtotal(self):
        return self.quantitat * self.preu

    def __str__(self):
        return f"Producte: {self.id_producte} - {self.quantitat} x {self.preu} €"


class Comanda:
    def __init__(self, id_comanda, data):
        self.id = id_comanda
        self.data = data
        self.items = []

    def afegir_item(self, item):
        self.items.append(item)

    def total(self):
        return sum(item.subtotal() for item in self.items)

    def __str__(self):
        return f"Comanda {self.id} ({self.data}) - Total: {self.total():.2f} €"


class User:
    def __init__(self, nom, edat, ciutat, telefons, localitzacio, etiquetes):
        self.nom = nom
        self.edat = edat
        self.ciutat = ciutat
        self.telefons = telefons
        self.localitzacio = localitzacio
        self.etiquetes = etiquetes
        self.comandes = []

    def afegir_comanda(self, comanda):
        self.comandes.append(comanda)

    def total_comandes(self):
        return sum(comanda.total() for comanda in self.comandes)

    def __str__(self):
        return f"{self.nom} ({self.ciutat})"
```

```python
users = []

# John
john = User(
    nom="John",
    edat=30,
    ciutat="New York",
    telefons=["+1 111 111 111", "+1 222 222 222"],
    localitzacio=(40.7128, -74.0060),
    etiquetes={"admin", "premium"}
)

comanda1 = Comanda(2222, "2023-07-01")

comanda1.afegir_item(ItemComanda(333, 3, 1.55))
comanda1.afegir_item(ItemComanda(555, 2, 2.55))

john.afegir_comanda(comanda1)

users.append(john)

# Anna
anna = User(
    nom="Anna",
    edat=25,
    ciutat="London",
    telefons=["+44 111 111 111"],
    localitzacio=(51.5074, -0.1278),
    etiquetes={"user", "active"}
)

users.append(anna)

# Carlos
carlos = User(
    nom="Carlos",
    edat=35,
    ciutat="Barcelona",
    telefons=["+34 600 123 456", "+34 600 987 654"],
    localitzacio=(41.3851, 2.1734),
    etiquetes={"premium", "editor"}
)

users.append(carlos)

# Emma
emma = User(
    nom="Emma",
    edat=28,
    ciutat="Paris",
    telefons=["+33 111 222 333"],
    localitzacio=(48.8566, 2.3522),
    etiquetes={"guest"}
)

users.append(emma)
```
