**lambda**

**Funcions i lambdas**

```python
def square(x):
  return x * x



print(square(4))

square_lamda = lambda x: x * x

print(square_lamda(5))
```

**Sense LAMBDAS**

```python
fresh_products = [("Apple", 1.2, 100), ("milk", 2.5, 200), ("Cheese", 5.0,600)]

def convert_to_dollars(p):
    return (p[0], p[1] * 1.1)

price_in_dollars = list(map(convert_to_dollars, fresh_products))

print(price_in_dollars)
```

Amb LAMBDAS

```python

fresh_products = [("Apple", 1.2, 100), ("milk", 2.5, 200), ("Cheese", 5.0,600)]

price_in_dollars = list(map(lambda p: (p[0], p[1] * 1.1), fresh_products))
print(price_in_dollars)
```

```python
expensive_products = list(filter(lambda p: p[1] > 2.0, fresh_products))

print(expensive_products)
```

```python
from functools import reduce

total_price = reduce (lambda acc, p:acc+p[1], fresh_products, 0)

print(total_price)
```

```python
products = [
    {"name": name, "price": price, "sales": sales}
    for name, price, sales in fresh_products
]

print(products)


# Ordenació per nom
sorted_by_name = sorted(products, key=lambda product: product["name"])

# Ordenació per preu
sorted_by_price = sorted(products, key=lambda product: product["price"])

# Ordenació per nombre de vendes
sorted_by_sales = sorted(products, key=lambda product: product["sales"])

print("Ordenat per nom:")
print(sorted_by_name)

print("Ordenat per preu:")
print(sorted_by_price)

print("Ordenat per vendes:")
print(sorted_by_sales)
```
